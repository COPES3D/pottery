```python
import os
import numpy as np
import h5py
from random import *
```

y가 클수록 바닥쪽  

label 1:  { M, 80, 7, -87, -M}  
label 2:  { M, 34, 13, -M}  
label 3:  { M, 71, 36, -M}  
label 4:  { M, 46, 2, -M}  
label 5:  { M, 128, 7, -148, -M}

```python
flist=os.listdir("./data/npy_10000/list")
max_shards=20
for file in flist:
    if 'filelist_1' in file: #file = filelist_1_shards_randomseed_F
        fstr=file.split('_') #{filelist, 1, shards, randomseed, F}
        filenames=[line.rstrip() for line in open("./data/npy_10000/list/"+file, 'r')] # filenames = generated_shards_0_randomseed_part_partnum
        a_data=np.zeros((len(filenames),2048,3))
        a_label=np.zeros((len(filenames),1),dtype=np.uint8)
        a_seg=np.zeros((len(filenames),4), dtype=np.uint8)
        temp_list=[]
        temp_label=[]
        temp_seg=[]        
        for i in range(0,len(filenames)):
            if filenames[i]=="":
                continue
            npydata=np.load("./data/npy_10000/one/"+filenames[i]+".npy")
            np.random.shuffle(npydata)
            if npydata.shape[0]<2048:
                continue
            else:
                temp_j_list=[]
                for j in range(0,2048):
                    temp_j_list.append([npydata[j][0],npydata[j][1], npydata[j][2]])
                temp_j_seg=[0,0,0,0]
                for k in range(len(npydata[:,1])):
                    seg=[0,0,0,0]
                    if npydata[k,1] >= 80:
                        seg[0] = 1
                    elif npydata[k,1] >= 7:
                        seg[1] = 1
                    elif npydata[k,1] >= -87:
                        seg[2] = 1
                    elif npydata[k,1] < -87:
                        seg[3] = 1
                    for j in range(len(seg)):
                        if seg[j] > temp_j_seg[j]:
                            temp_j_seg[j]=seg[j]
                        else:
                            temp_j_seg[j]=temp_j_seg[j]    
                temp_list.append(temp_j_list)
                temp_label.append(int(fstr[1])-1)
                temp_seg.append(temp_j_seg)
                
        if len(temp_list) < max_shards:
            for i in range(max_shards-len(temp_list)):
                temp_shuffle_list=temp_list
                temp_shuffle_seg=temp_seg
                temp_list.append(temp_shuffle_list[randrange(len(temp_shuffle_list))])
                temp_label.append(int(fstr[1])-1)
                temp_seg.append(temp_shuffle_seg[randrange(len(temp_shuffle_seg))])

            
        a_data=np.array(temp_list)
        a_label=np.transpose(np.array(temp_label))
        a_seg=np.array(temp_seg)
        
        target=np.zeros(a_data.shape)
        for i in range(a_data.shape[0]):
            for j in range(a_data.shape[1]):
                for k in range(a_data.shape[2]):
                    target[i][j][k]=a_data[i][j][k]-a_data[i].mean(axis=0)[k]

        f=h5py.File("./data/npy_10000/h5/"+str(fstr[1])+'_'+str(fstr[2])+'_'+str(fstr[3])+".h5",'w')
        data=f.create_dataset("data", data=target)
        label=f.create_dataset("label", data=a_label)
        seg=f.create_dataset("seg", data=a_seg)
        f.close()
```
