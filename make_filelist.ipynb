```python
folders_one=os.listdir("./data/npy_10000/one")
folders_two=os.listdir("./data/npy_10000/two")
folders_three=os.listdir("./data/npy_10000/three")
folders_four=os.listdir("./data/npy_10000/four")
folders_five=os.listdir("./data/npy_10000/five")
folders_one.sort()
folders_two.sort()
folders_three.sort()
folders_four.sort()
folders_five.sort()
```

```python
# 한 label 별 filelist 만들기
# filelist_label_shards_randomseed.txt 만들기
for item in folders_one:
    if 'npy' in item:
        for n in range(5,21):
            if 'generated_'+str(n) in item:
                if "part_5" in item:
                    fstr=item.split('_')
                    filelist=open('./data/npy_10000/filelist_'+str(1)+'_'+str(fstr[1])+'_'+str(fstr[3])+'_F','w')
                    filelist.close()
```

```python
# 위에서 만든 filelist 열고 안에 .npy 파일명들 넣기
folder=os.listdir("./data/npy_10000")
folder.sort()
for text in folder:
    if 'filelist' in text:
#         print(text)
        fstr=text.split('_')
#         print(fstr)
        if str('_'+fstr[3]+'_F') in text: #fstr[3]=randomseed
#             print(text)
            f=open("./data/npy_10000/"+text,"w")
            for item in folders_one:
                if 'npy' in item:
                    if str('_'+fstr[3]+'_') in item:
                        print(item)
                        filename=item[:-4]
#                         print(filename)
                        f.write("%s\n" % filename)
            f.close()
```
